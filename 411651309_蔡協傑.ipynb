```python
score = int(input("score: "))
if score >= 90:
    print('Grade is A')
elif score >= 80:
    print('Grade is B')
elif score >= 70:
    print('Grade is C')
elif score >= 60:
    print('Grade is D')
else:
    print('Grade is F')
```

```python
a=int(input('請輸入體溫'))
if a>=37.5:
  print('溫度過高')
elif 37<=a<=37.4:
  print('溫度略高')
elif 34<=a<=36.9:
  print('體溫正常')
else:
  print('體溫異常,請重新量測')
```

```python
a=int(input('請輸入月份'))
if 1<=a<=12:
  if 3<=a<=5:
    print(a,'月是春天')
  elif 6<=a<=8:
    print(a,'月是夏天')
  elif 9<=a<=11:
    print(a,'月是秋天')
  else:
    print(a,'月是冬天')
else:
  print('月份不在範圍內')
```

```python
a=int(input('請輸入今年收入淨額'))
if a>300000:
  if a>=2000000:
    print('賦稅金額',a*0.3,'元')
  elif a>1000000:
    print('賦稅金額',a*0.21,'元')
  elif a>600000:
    print('賦稅金額',a*0.13,'元')
  else:
    print('賦稅金額',a*0.06,'元')
else:
  print('賦稅金額:0元')
```
